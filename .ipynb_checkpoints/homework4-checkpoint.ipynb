```python
#1

import pandas as pd

df = pd.read_csv("./data/sea.csv")

from skmultiflow.data.waveform_generator import WaveformGenerator
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from skmultiflow.data.data_stream import DataStream

stream = WaveformGenerator()

stream = DataStream(data = df)
stream.prepare_for_use()
clf = SVC(gamma=0.01)

stream.restart()
X, y = stream.next_sample(10000)

clf.fit(X, y)

print(clf.score(X, y))
    
X_test, y_test = stream.next_sample(90000)

print(clf.score(X_test,y_test))

# train the first 10000 samples of the stream, than evaluate with train data and test data/remaining data
    
```

```python
#2

def reset_values():
    ph_mean = 0
    ph_sum = 0
    ph_sum_list = []
    sample_count = 1
    return(ph_mean, ph_sum, ph_sum_list, sample_count)

def compute_ph_statistics(mean_score, ph_sum, score, ph_sum_list):
    # Compute new mean
    mean_new = mean_score + ((score - mean_score) / sample_count)
    
    # Compute new sum of deviations
    ph_sum_new = ph_sum + (score - mean_new)
    ph_sum_list.append(ph_sum_new)
    
    # Compute change statistics
    statistic_max = max(ph_sum_list) - ph_sum_new
    statistic_min = ph_sum_new - min(ph_sum_list)
    
    return (mean_new, ph_sum_new, ph_sum_list, statistic_max, statistic_min)


stream.restart()
clf_two = SVC(gamma=0.01)

# Set initial values
sample_count = 1
ph_mean = 0
ph_sum = 0

# Threshold for change detection
ph_lambda = 1

ph_sum_list = []
ph_score_list = []

X_train, y_train = stream.next_sample(10000)
clf_two.fit(X_train, y_train)

for i in range(300):
    X_test, y_test = stream.next_sample(100)
    stream.n_remaining_samples()
    score = clf_two.score(X_test, y_test)
    ph_score_list.append(score)
    
    ph_mean, ph_sum, ph_sum_list, statistic_max, statistic_min = compute_ph_statistics(ph_mean, ph_sum, score, ph_sum_list)
    sample_count += 1

    if ((statistic_max > ph_lambda) | (statistic_min > ph_lambda)):
        ph_mean, ph_sum, ph_sum_list, sample_count = reset_values()
        print('Drift, No. of iterations:', i)
        clf_two.fit(X_test, y_test)

        
plt.figure()
plt.plot(ph_score_list)
plt.show()

# implement PH-Lambda like Lucas Baier in excercise 5  - output oft drifts and plot

```

```python
#3

def reset_values():
    ph_mean = 0
    ph_sum = 0
    ph_sum_list = []
    sample_count = 1
    return(ph_mean, ph_sum, ph_sum_list, sample_count)

def compute_ph_statistics(mean_score, ph_sum, score, ph_sum_list):
    # Compute new mean
    mean_new = mean_score + ((score - mean_score) / sample_count)
    
    # Compute new sum of deviations
    ph_sum_new = ph_sum + (score - mean_new)
    ph_sum_list.append(ph_sum_new)
    
    # Compute change statistics
    statistic_max = max(ph_sum_list) - ph_sum_new
    statistic_min = ph_sum_new - min(ph_sum_list)
    
    return (mean_new, ph_sum_new, ph_sum_list, statistic_max, statistic_min)

stream.restart()

# parammeters for gridsearch gamma for SVM and Ph_lambdas
param_gamma = [0.01, 0.1, 1]
             
param_ph = [0.01, 0.05, 0.1]


X_train, y_train = stream.next_sample(10000)

max_score = 0
best_parameter_str = ''

for gamma in param_gamma:
    clf_three = SVC(gamma=gamma)
    clf_three.fit(X_train, y_train)
    
    # Set initial values
    sample_count = 1
    ph_mean = 0
    ph_sum = 0

    # Threshold for change detection
    ph_sum_list = []
    ph_score_list = []
    
    for ph_lambda in param_ph:
        X_test, y_test = stream.next_sample(1000)
        score = clf_three.score(X_test, y_test)
        ph_score_list.append(score)
    
        ph_mean, ph_sum, ph_sum_list, statistic_max, statistic_min = compute_ph_statistics(ph_mean, ph_sum, score, ph_sum_list)
        sample_count += 1

        if ((statistic_max > ph_lambda) | (statistic_min > ph_lambda)):
            ph_mean, ph_sum, ph_sum_list, sample_count = reset_values()
            clf_four = SVC(gamma=gamma)
            clf_four.fit(X_test, y_test)
            score = clf_four.score(X_test, y_test)
        
        if(score > max_score):
            max_score = score
            best_parameter_str = 'gamma:' + str(gamma) + ' phLambda:' + str(ph_lambda) + ' score:' + str(score)
        
        print('gamma:' + str(gamma) + ' phLambda:' + str(ph_lambda) + ' score:' + str(score))
        
print('Pair of parameters with highest score: ' + best_parameter_str)

# using a nested loop to test all combination of parameters - output of gamma and lambda pair with score
# print best score at the ende again
```
